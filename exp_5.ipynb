```python
#Build a simple MLP and implement forward propagation and Train the MLP on a small dataset


!pip install openpyxl

from google.colab import files
uploaded=files.upload()

from google.colab import drive
drive.mount('/content/drive')

!ls /content/drive/MyDrive/

import pandas as pd
# Creating a dataset
dataset_path= "/content/drive/MyDrive/ANN/Student_dataset.xlsx"

df=pd.read_excel(dataset_path,engine='openpyxl')
print(df.head())

import numpy as np
import pandas as ps
#Initailizing Parameters
def initialize_parameters(layer_dims):
  np.random.seed(3)
  print("Layer Dimensions:", layer_dims)
  parameters={}
  L = len(layer_dims)
  print("Total No. of layers in NN",L)
  for i in range(1, L):
      parameters['W' + str(i)] = np.ones((layer_dims[i-1],
layer_dims[i])) * 0.1
      parameters['b' + str(i)] = np.zeros((1, layer_dims[i])) # Modified Line

  return parameters
layer_dims=[4,2,1]
parameters=initialize_parameters(layer_dims)
print(parameters)
def linear_forward(A_prev, W, b):
    Z = np.dot(A_prev, W) + b
    return Z
def relu(Z):

    A = np.maximum(0, Z)
    return A

def L_layer_forward(X, parameters):

    caches = []
    A = X
    L = len(parameters) // 2                  # number of layers in the neural network

    # Implement [LINEAR -> RELU]*(L-1). Add "cache" to the "caches" list.
    for l in range(1, L):
        A_prev = A
        # Retrieve parameters from the dictionary "parameters"
        W = parameters['W' + str(l)]
        b = parameters['b' + str(l)]
        #print("A_prev shape:", A_prev.shape)
        #print("W shape:", W.shape)
        #print("b shape:", b.shape)
        # Linear forward calculation
        Z = linear_forward(A_prev, W, b)
        # Relu activation function
        A = relu(Z) #Calling the relu function which was missing before
        # Store the cache for backpropagation
        cache = (A_prev, W, b, Z)
        caches.append(cache)

    # Implement LINEAR -> SIGMOID. Add "cache" to the "caches" list.
    # Retrieve parameters for the output layer
    WL = parameters['W' + str(L)]
    bL = parameters['b' + str(L)]

    # Linear forward calculation for the output layer
    #print("A shape:", A.shape)
    #print("WL shape:", WL.shape)
    #print("bL shape:", bL.shape)

    ZL = linear_forward(A, WL, bL)
    AL = ZL   # In this case, the activation is just the linear output

    # Store the cache for backpropagation (output layer)
    cache = (A, WL, bL, ZL)
    caches.append(cache)

    return AL, caches

#Example execution
import numpy as np
import pandas as pd


# Prepare input (X) and output (Y) data from the dataset
X = df[['CGPA', '10th Score' ,'12th Score' ,'IQ']].values  # Features: Hours and Attendance
Y = df['Placement'].values  # Target variable: Marks

# Reshape Y to match the expected shape for L_layer_forward
Y = Y.reshape(Y.shape[0], 1)

# Example execution
# Call L_layer_forward to get the output
output, _ = L_layer_forward(X, parameters)

# Now you can print the output
print("Output (y):", output)
```
